# 3. Leading space file problems

```python
# Fix some leading space problems
import os
import glob
# all_audio_files = glob.glob('output/tts/en-US-AvaMultilingualNeural/*')
all_audio_files = glob.glob('output/tts/zh-CN-XiaoyuMultilingualNeural/*')
print(len(all_audio_files), all_audio_files[:3])
for i_file, audio_file in enumerate(all_audio_files):
    if audio_file.split('/')[-1][0] == ' ':
        new_audio_file = '/'.join(audio_file.split('/')[:-1]) + '/' + audio_file.split('/')[-1].strip()
        os.rename(audio_file, new_audio_file)
        print(audio_file)
        print(new_audio_file)
    else:
        print(f'do not rename {audio_file}')
```

# 7. Add hsk column to google sheet

```python
sheet_url = 'https://docs.google.com/spreadsheets/d/1pw9EAIvtiWenPDBFBIf7pwTh0FvIbIR0c3mY5gJwlDk/edit#gid=0'
sheet_url = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(sheet_url).sort_values('id')
df.head()
```

```python
# Import HSK words and save into their own combined csv
hsk_raw_files = sorted(glob.glob('static/hsk/HSK*.tsv'))
dfs_all = []
for hsk_file in hsk_raw_files:
    df_tmp = pd.read_csv(hsk_file, delimiter='\t', header=None).rename(columns={0:'trad', 1:'chinese', 2:'hsk3.0_pinyin', 3: 'hsk3.0_def'})
    df_tmp['hsk_level'] = hsk_file.split(' ')[-1][:-4]
    dfs_all.append(df_tmp)
df_all_hsk = pd.concat(dfs_all).reset_index(drop=True)
df_all_hsk.to_csv('static/hsk/hsk_1to9.csv')
print(len(df_all_hsk))
df_all_hsk.head()
```

```python
# Drop duplicates for hsk
df_all_hsk_nodup = df_all_hsk.drop_duplicates('chinese')[['chinese', 'hsk_level']]
df_mmd_hsklevels = df[['id', 'chinese']].merge(df_all_hsk_nodup, on='chinese', how='left').fillna('NONE')
df_mmd_hsklevels.to_csv('static/hsk/mmd_hsk_20251002.csv')
df_mmd_hsklevels[['hsk_level']].to_csv('static/hsk/mmd_hsk_20251002_tmp.csv', index=False)
```
